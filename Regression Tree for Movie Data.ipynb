```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Movie_regression.csv')
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

#### Missing value imputation

```python
df.Time_taken.mean()
```

```python
df['Time_taken'].fillna(value=df.Time_taken.mean(),inplace=True)
```

```python
df.isna().sum()
```

### Dummy Variable creation

```python
df=pd.get_dummies(df,columns=['3D_available','Genre'],drop_first=True)
```

```python
df.describe()
```

#### x-y splitting of the variables.

```python
x=df.loc[:,df.columns!='Collection']
```

```python
type(x)
```

```python
y=df[['Collection']]
type(y)
```

```python
y.head()
```

#### Train_Test_Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

### Training a regression tree

```python
from sklearn import tree
```

```python
regtree=tree.DecisionTreeRegressor(max_depth=3)
regtree.fit(x_train,y_train)
```

```python
### predict values
y_test_pred=regtree.predict(x_test)
y_train_pred=regtree.predict(x_train)
```

### Evaluation performance

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mean_squared_error(y_test,y_test_pred)
```

```python
r2_score(y_train,y_train_pred)
```

```python
r2_score(y_test,y_test_pred)
```

> Note: R^2 value between 0.4 and 1 is good for a model to fit. 

### plotting regression tree

```python
dot_data=tree.export_graphviz(regtree,out_file=None,feature_names=x.columns,filled=True)
```

```python
import graphviz
graphviz.Source(dot_data)
```

```python
x_train.columns
```

### Pruning Tree-controlling tree growth

#### Maximum number of levels in tree

```python
regtree1 = tree.DecisionTreeRegressor(max_depth = 3)
regtree1.fit(x_train, y_train)
dot_data1 = tree.export_graphviz(regtree1, out_file=None,feature_names= x_train.columns, filled = True)
graphviz.Source(dot_data1)
```

#### Minimum observations at internal node

```python
regtree2 = tree.DecisionTreeRegressor(min_samples_split = 40,max_depth=4)
regtree2.fit(x_train, y_train)
dot_data2 = tree.export_graphviz(regtree2, out_file=None,feature_names= x_train.columns, filled = True)
graphviz.Source(dot_data2)
```

#### Minimum observations at leaf node

```python
regtree3 = tree.DecisionTreeRegressor(min_samples_leaf = 25)
regtree3.fit(x_train, y_train)
dot_data3 = tree.export_graphviz(regtree3, out_file=None,feature_names= x_train.columns, filled = True)
graphviz.Source(dot_data3)
print(r2_score(y_test,regtree3.predict(x_test)))
```

> On changing the decision tree parameters 
>- which are max_depth,min_samples_split,min_samples_leaf 
>- we can obtain the desired controlled tree and the r2_score can be varied and from the r2_score we can tell the accuracy of the model
